**CO2 Emission Prediction**

Problem: Predict CO2 emissions from cars based on engine size, fuel consumption, and cylinders.

Dataset Link: https://www.kaggle.com/datasets/debajyotipodder/co2-emission-by-vehicles?resource=download

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/CO2 Emissions_Canada.csv')
data
```

```python
training_set = data.drop(['Make', 'Model', 'Vehicle Class','Transmission','Fuel Type','Fuel Consumption City (L/100 km)','Fuel Consumption Hwy (L/100 km)','Fuel Consumption Comb (mpg)'], axis=1)
training_set
```

```python
features = training_set[['Engine Size(L)','Cylinders','Fuel Consumption Comb (L/100 km)']]
target = training_set['CO2 Emissions(g/km)']

#Scale the features
mean = np.mean(features, axis=0)
std_dev = np.std(features, axis = 0) #Standard deviation measures the amount of variation or dispersion of a set of values.

#Standardization
x_scaled = (features - mean)/std_dev
```

```python
#We have three features then we need three weights.
num_features = x_scaled.shape[1]
w = np.zeros(num_features)
b = 0.0

#Hyperparameters
learning_rate = 0.01
epochs = 1000 #number of iterations
n_sample = len(target)
```

```python
#Gradient Descent
for i in range(epochs):
  prediction = np.dot(x_scaled,w)+b

  error = prediction - target

  #Gradient calculation
  dw = (1/n_sample)*np.dot(x_scaled.T,error)
  db = (1/n_sample)*np.sum(error)

  #update parameter
  w -= learning_rate*dw
  b -= learning_rate*db

print(f"Learned Weights: {w}")
print(f"Learned Bias (Intercept): {b:.4f}")
```

```python
#make prediction
new_car_features = np.array([3.5, 6, 10.6])

# IMPORTANT: must scale the new data using the SAME mean and std_dev from the training data
new_car_scaled = (new_car_features - mean) / std_dev

# Make the prediction
predicted_co2 = np.dot(new_car_scaled, w) + b

print(f"\nPrediction for the new car: {predicted_co2:.2f} g/km")
```

```python
"""
Calculates the Mean Absolute Percentage Error (MAPE).

Args:
    y_true (np.array): The true values.
    y_pred (np.array): The predicted values.

Returns:
    float: The MAPE value as a percentage.
"""
# To avoid division by zero, we add a small epsilon
# This is a standard practice for MAPE to handle cases where true values are zero.
epsilon = 1e-10
mape = np.mean(np.abs((target - prediction) / (target + epsilon))) * 100
print(mape)
```
